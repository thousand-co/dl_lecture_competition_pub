```python
#import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#from PIL import Image
import os
import glob

imfolder='.\\Images\\'
```

```python
f=glob.glob(os.path.join(imfolder, '**\\*'))
df_f=pd.DataFrame(f)
df_f.columns=['w_path']
df_f
```

```python
df_f['folder']=df_f.w_path.map(lambda x: x.split('\\')[2])
df_f['fname']=df_f.w_path.map(lambda x: x.split('\\')[3])
df_f
```

```python
len(df_f.fname.unique())
```

```python
fn_dict=dict(zip(df_f.fname, df_f.folder))
fn_dict
```

```python
df1=pd.read_csv('./data/train_image_paths.txt', header=None)
df1.columns=['fpath']
df1
```

```python
df1['folder']=df1.fpath.map(lambda x: x.split('/')[0] if len(x.split('/'))>1 else '')
df1['fname']=df1.fpath.map(lambda x: x.split('/')[1] if len(x.split('/'))>1 else x.split('/')[0])
df1['folder']=df1.fname.map(lambda x: fn_dict[x])
df1['fpath2']=df1.folder+'/'+df1.fname
for i, raw in df1.iterrows():
    if raw.fpath!=raw.fpath2:
        df1.loc[i, 'check']=1
df1
```

```python
df1.fpath2.to_csv('./data/train_image_paths_rev.txt', header=False, index=False)
```

```python
df1[df1['check']==1].shape
```
